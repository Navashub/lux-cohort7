# TUPLES
  - a tuple is just like a list, but immutable - once created , you cannot change it.
  - they are faster than list and you use tuples when data should stay fixed

  Examples uses: Coordinates(lat,lng) , RGB colors (255,0,0), days of the week

  - use ( ) to create tuples

## Creating and Using Tuples

```python
# use round brackets ( )
point = (3,7)
colors = ("red", "green", "blue")
person = ("Bob", 25, "Nairobi")

# A single-item tuple NEEDS a trailing comma
singe = (42,) # without the comma it is just (42)

print(point)
print(colors)
print(person)
```

### Access items - same as list

```python
colors = ("red", "green", "blue")
#          0       1        2
print(colors[0]) # red
print(colors[-1]) # blue
print(colors[1:]) # green and blue
```

```python
colors[0] = "yellow" # TypeError: 'tuple' object does not support item assignment
```

## Tuple Methods
- only have 2 methods because they cannot be changed
- count() - Returns the number of times a specific value occurs in the tuple.
- index() - Searches the tuple for a value and returns the index of its first occurrence.

```python
nums = (1,2,3,2,2,5)

print(nums.count(2)) # 3 - how many times 2 appears
print(nums.index(2)) # 1 -  position of the first 2
print(nums.index(1))
print(nums.index(5))
```

### Tuples Unpacking
  - you can unpack a tuple into individual variables in one line

```python
point = (10,20)
x, y = point
print(x)
print(y)
```

```python
# swap two variables using a tuple
a = 5
b = 10

a,b = b,a
print(a,b)
```

```python
# a function can return a tuple
def min_max(numbers):
  return min(numbers), max(numbers)

low,high = min_max([3,1,9,4,7])
print(low, high)
```

# SETS
  - A set is unordered collection of UNIQUE values.
  - Duplicates are automatically re,oved.

- sets are useful for removing duplicates from a list and for comparing group of items

### Creating Sets
- we use { } (curly braces)

```python
fruits = {"apple", "banana", "mango", "apple", "mango"}
print(fruits) # {'banana', 'mango', 'apple'} - "apple", "mango" - appear only once!
```

```python
# IMPORTANT: Empty set must use set() NOT {}

empty = set() # correct
empty = {} # WRONG - this creates an empty dictionary
```

```python
# remove duplicates from a list usng set
nums = [1,2,2,3,3,3,4,5,5]
unique_nums = set(nums)
print(unique_nums)
```

```python
# Add and remove
#add
fruits = {'banana', 'mango', 'apple'}
fruits.add("cherry")
print(fruits)
```

```python
fruits.remove("banana") # removes (error if not found)
print(fruits)
```

```python
fruits.discard("kiwi")  # removes (NO error if not found)
print(fruits)
```

## SET OPERATIONS
  - sets does well when you want to compare two groups

```python
class_a = {"Rita", "Idah", "Jullian", "Bob", "Abiud"}
class_b = {"Bob", "David", "Timothy", "Idah"}

# Union - everyone from both classes ( | )
# "Rita", "Idah", "Jullian", "Bob", "Abiud","David","Timothy"
print(class_a | class_b)
```

### Intersection - Only elements present in both sets. ( & )
    - students in both classes

```python
class_a = {"Rita", "Idah", "Jullian", "Bob", "Abiud"}
class_b = {"Bob", "David", "Timothy", "Idah"}

# "Idah", "Bob"
print(class_a & class_b)
```

## Difference - Elements in the first set but not the second. ( - )
  - in class_a but NOT in class_b

```python
class_a = {"Rita", "Idah", "Jullian", "Bob", "Abiud"}
class_b = {"Bob", "David", "Timothy", "Idah"}

# "Rita", "Jullian", "Abiud"
print(class_a - class_b)
```

### Symmetric Difference - Elements in either set, but not both.

```python
class_a = {"Rita", "Idah", "Jullian", "Bob", "Abiud"}
class_b = {"Bob", "David", "Timothy", "Idah"}

# Rita", "Jullian", "Abiud", "David", "Timothy"
print(class_a ^ class_b)
```

```python
# you can also write this as methods
class_a.union(class_b)
class_a.intersection(class_b)
class_a.difference(class_b)
class_a.symmetric_difference(class_b)
```
