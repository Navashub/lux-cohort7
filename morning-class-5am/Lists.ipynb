# Accesing items - indexing
- Every item in a list has a number called an index.
- In python, counting starts from 0, not 1

```python
fruits = ["apple", "banana", "mango", "orange"]
#         index 0   index 1   index 2   index 3

print(fruits[0]) # apple
print(fruits[1]) # banana
print(fruits[2]) # mango
```

### Negative indexing
  - count from the end

```python
fruits = ["apple", "banana", "mango", "orange"]
print(fruits[-1]) # orange (last item)
print(fruits[-2]) # monago (second to last)
```

## List methods - things you can do with a list
  - Python gives lists built-in methods(functions you call on a list)

```python
fruits = ["apple", "banana", "mango", "orange"]
print(fruits)
```

# Add items

```python
# append() method
fruits.append("cherry") # adds at the END
print(fruits)
```

## add at specific position

```python
# add pear to position 1 - insert(position, item)
fruits.insert(1, "pear") # adds at position 1
print(fruits)
```

## Remove items

```python
# remove banana - remove()
fruits.remove("banana") # removes by value
print(fruits)
```

```python
# remove last item
fruits.pop()
print(fruits)
```

```python
# remove item at a specific index
fruits.pop(0)
print(fruits)
```

## Other useful methods

```python
# apple , banana, cherry
fruits.append("banana")
fruits.append("apple")
fruits.insert(0,"cherry")
print(fruits)

```

```python
# Sort aphabetically
fruits.sort()
print(fruits)
```

```python
# reversing a list
fruits.reverse()
print(fruits)
```

```python
# find position of an item
print(fruits.index("banana"))
```

```python
# count how many times an item appears

print(fruits.count("banana"))
```

## List Slicing - Getting a portion of a list

```python
nums = [10,20,30,40,50]
#       0   1  2  3  4

# from index 1 upto (not including) 3
print(nums[1:3])

# from start up to index 2
print(nums[:2])

# from index 2 to the end
print(nums[2:])

# reversed
print(nums[::-1])
```

# List Comprehension
 	-concise way to create lists in python
 	-compact and easier to read than the traditional loops

```python
# The long way (using a for loop)
# Create a list of squares: [1,4,9,16,25]

squares = []
for x in range(1,6): # 1, 2, 3,4 ,5
  squares.append(x ** 2)
print(squares)
```

```python
# The short way (list comprehension)

# Pattern : [expression for item in iterable if condition]
#           [what to do]   [where to get items] [optional filter]

squares = [x ** 2 for x in range(1,6)]
print(squares)
```

```python
# Pattern : [expression for item in iterable if condition]

# Only even number from 0-9
evens = [x for x in range(10) if x % 2 == 0]
print(evens)

# only odd numbers from 1-15
odds=[x for x in range(1,16) if x  % 2 != 0]
print(odds)

odds1=[x for x in range(1,16) if x  % 2 == 1]
print(odds1)
```

```python
# Pattern : [expression for item in iterable if condition]

# Uppercase all fruits
fruits = ["apple", "banana", "mango"]

upper_fruits = [f.upper() for f in fruits]
print(upper_fruits)
```
