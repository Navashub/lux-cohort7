# Data Pipeline Demo

This notebook shows how to fetch raw JSON data, clean it up, and save the first 100 rows as a CSV file.

## Step 1: Install / Import Libraries

Import the two libraries we need for this pipeline: `requests` to fetch the data and `pandas` to work with the table.

```python
import requests
import pandas as pd
```

## Step 2: Fetch the Data

Use the raw GitHub URL to download the JSON data. We check the status code and print a helpful error message if the request fails.

```python
url = 'https://raw.githubusercontent.com/Navashub/lux-cohort7/main/mockaroo-csv/users_data.json'
response = requests.get(url)

if response.status_code != 200:
    print(f'Error fetching data: HTTP {response.status_code}')
else:
    print('Data fetched successfully.')
```

## Step 3: Parse the JSON

Convert the downloaded JSON text into a Python list of dictionaries, then print the number of records to confirm the data loaded correctly.

```python
data = response.json()

print(f'Total records fetched: {len(data)}')
```

## Step 4: Trim to 100 Rows

Keep only the first 100 rows so the output file is a smaller, manageable example dataset.

```python
trimmed_data = data[:100]
print(f'Remaining records after trim: {len(trimmed_data)}')
```

## Step 5: Load into a pandas DataFrame

Turn the list of dictionaries into a DataFrame so we can inspect and export the data easily.

```python
df = pd.DataFrame(trimmed_data)
print('DataFrame created successfully.')
```

## Step 6: Inspect the DataFrame

Look at the first rows, the size of the DataFrame, and the type of each column.

```python
print('First 5 rows:')
display(df.head())

print('Shape:')
print(df.shape)

print('Data types:')
print(df.dtypes)
```

## Step 7: Export to CSV

Save the trimmed DataFrame to a CSV file with `index=False` so row numbers are not included.

```python
output_filename = 'users_data_100.csv'
df.to_csv(output_filename, index=False)
print(f'CSV exported successfully to {output_filename}')
```
